```python
import pandas as pd
import numpy as np
```

```python
books_df = pd.read_csv('book_recommondation.csv',encoding='latin-1')
books_df
```

```python
# remove the unnamed:0 column as it is containing just the serials numbers which is not required
del books_df['Unnamed: 0']
```

```python
books_df
```

```python
#number of unique users in the dataset
len(books_df['User.ID'].unique())
```

```python
#number of unique books in the given dataset
len(books_df['Book.Title'].unique())
```

```python
# As there are large number of unique users and books for the given dataset
# To ensure the Statistical Significance,users with less than 50 rating and books with less than
#100 ratings are excluded
# As collaberative filtering 

# filtering the users based on the number of ratings they have given 
counts1 = books_df['User.ID'].value_counts()
books_df = books_df[books_df['User.ID'].isin(counts1[counts1>=50].index)]

# filtering the books based on the ratings given to it(un popular books)
counts = books_df['Book.Rating'].value_counts()
books_df = books_df[books_df['Book.Rating'].isin(counts[counts>=150].index)]
```

```python
books_df
```

```python
len(books_df['User.ID'].unique())
```

```python
#number of unique books in the given dataset
len(books_df['Book.Title'].unique())
```

```python
books_df
```

```python
users_books_df= books_df.reset_index().pivot_table(index='User.ID',
                                 columns='Book.Title',
                                 values='Book.Rating')
```

```python
users_books_df
```

```python
users_books_df.index = books_df['User.ID'].unique()
users_books_df
```

```python
#Impute those NaNs with 0 values
users_books_df.fillna(0, inplace=True)
```

```python
users_books_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
# pairwise distance will caluclate the distance b/w one row to every other row
# when we say pairwise distance it gives one - cosine based distance to get the normal cosine distance 
#we convert that it into one - pairwise
user_sim = 1 - pairwise_distances(users_books_df.values,metric='cosine')
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books_df['User.ID'].unique()
user_sim_df.columns = books_df['User.ID'].unique()
```

```python
user_sim_df.iloc[0:5, :]
```

```python
# fill all the diagnol values with zero as we are searching for most similar user
# it will give the same user details
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, :]
```

```python
#most similar users for first 5 users
user_sim_df.idxmax(axis=1)[1:5]
```
